Consider an organization which executes website development projects for their clients. Employees are employed on these projects. The data about *Project titles*, *Project IDs*, *Employees* and *Employees working on Projects* are provided in a text file. The first $10$ entries in the given text file are Project Titles. Next $6$ entries are the name of the employees.

`projects.txt`

Project Screenshot.png

Project IDs are given in next $10$ records and last $6$ entries are strings of $10$ characters with $0$'s and $1$'s. One such string defines the assignment of the project by setting $1$ at the positional index representing a project.

For example $0010010101$- defines that Rita Hoggins works on  projects â€“ Neil Boutique(Project ID $1003$), Creative Fingers Eating House(Project ID $3001$), Helping Hands Grocery(Project ID  $3003$), Smirk Smile Corrector(Project ID $4002$).

Return a dictionary the maps the employee names with a dictionary mapping the project ids with their names.

```python
# Reading the file
with open('Project.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
type(content)
```

```python
# Split the string by new line
content.split('\n')
```

```python
# Save elements in each line into a list
file_content_list = content.split('\n')
```

```python
# Slice the project names
projects = file_content_list[:10]
projects
```

```python
# Slice the employee names
employees = file_content_list[10: 16]
employees
```

```python
# Slice the project IDs
project_ids = file_content_list[16: 26]
project_ids
```

```python
# Slice the employee-project mappers
project_emp_mapper = file_content_list[26:]
project_emp_mapper
```

```python
# Create id-project mapping dictionary
id_project_map = []
for i in range(10):
    id_project_map.append({project_ids[i]: projects[i]})

id_project_map
```

```python
# Returns the a log which maps employee names with a dictionary containing projects assigned to each employee
def map_project_emp(projects_log=None, employees=None, mapper_ids=None):
    project_employee_log = {}
    for i in range(len(project_emp_mapper)):
        project_employee_log[employees[i]] = {}
        for j in range(len(project_emp_mapper[i])):
            if project_emp_mapper[i][j] == '1':
                project_employee_log[employees[i]].update(projects_log[j])

    return project_employee_log
```

```python
project_employee_log = map_project_emp(id_project_map, employees, project_emp_mapper)
project_employee_log
```
